## Cantidad de kills influye en la victoria del equipo?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
db = 'https://raw.githubusercontent.com/razbackup/CSV-EV2/main/dbcs.csv'
data = pd.read_csv(db, engine='python', sep=';', encoding='utf-8')
```

```python
kills_by_team = data.groupby('Team')['MatchKills'].sum()
print("Total Kills por equipo:")
print(kills_by_team)

# Graficar la cantidad total de kills por equipo
sns.barplot(x=kills_by_team.index, y=kills_by_team.values)
plt.title('Total Kills por equipo')
plt.xlabel('Equipo')
plt.ylabel('Total Kills')
plt.show()
```
